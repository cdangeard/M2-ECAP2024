Récupérer tout les liens présents dans la page

```python
liens = soup.find_all('a', href=True, title=True)
liens
```

Conservont dans une liste uniquement les liens qui pointes vers une autre page wikipedia

```python
liensWiki = [lien.get('href') for lien in liens if lien.get('href').startswith('/wiki/')]
liensWiki
```

Récuperer le tableau des types en python et le stocker dans un dictionnaire (ou un dataframe)

```python
tableaux = soup.find_all('table', class_='wikitable')
print([tableau.caption.string for tableau in tableaux])
```

```python
# Dans notre cas, on pourait se contenter de la première table
# On peut aussi chercher parmi les tables celle qui nous intéresse
# et definir un critère de recherche
for table in tableaux:
    if table.caption.string == 'Summary of Python 3\'s built-in types\n':
        tableType = table
        print('table Found')
        break
```

```python
lignes = tableType.find_all('tr')
header = [cellule.text.strip() for cellule in lignes[0].find_all('th')]

tableDesTypes  = [
    {header[i]: cellule.text.strip() for i, cellule in enumerate(ligne.find_all('td'))}
    for ligne in lignes[1:]
]
```

```python
tableDesTypes
```

```python
df_typesPython = pd.DataFrame(tableDesTypes)
df_typesPython
```

# Bonus

- Quelle est la différence entre find et findChild?
- Quelle est la différence entre children et descendants ?
- Comment récupérer le contenu d'un attribut ?

```python
pageTest = """<!DOCTYPE html>
<html>
<head>
<title>Page de test</title>
</head>
<body>
<h1>Mon premier titre</h1>
<p>Mon premier paragraphe</p>
<h1>Mon deuxième titre</h1>
<p>Mon deuxième paragraphe avec un <a href="#lien">lien Interne</a></p>
</body>
</html>
"""

soupTest = BeautifulSoup(pageTest, 'html.parser')
```

```python
print(soupTest.findChild('h1', recursive=False)) # On ne cherche que
print(soupTest.find('h1')) # ctrl + clic pour voir la doc 
```

```python
print(f'nombre d\'enfants : {len([child for child in soupTest.children])}')
print(f'nombre de descendants : {len([descendant for descendant in soupTest.descendants])}')

# Donc findChild est mal nommé, il devrait s'appeler findDescendant
print(soupTest.findChild('h1', recursive=False)) # on cherche parmis les enfants
print(soupTest.findChild('h1')) # on cherche parmis les descendants
```

# SQLite

Créer et ajouter le tableau des types en python dans une base de donnée SQLite

Vérfier que le tableau est bien présent dans la base de donnée

# Documentation

https://docs.python.org/3/library/sqlite3.html

```python
import sqlite3

# On cree une connexion à la base de données (cas échéant, elle sera créée)
conn = sqlite3.connect('dbWebscrapping.db')
```

```python
# On enregistre le dataframe dans la base de données
df_typesPython.to_sql(name='typePython', con=conn, if_exists='replace', index=False)
```

```python
# On peut maintenant lire la table
pd.read_sql_query('SELECT * FROM typePython', conn)
```

```python
conn.close() # On ferme la connexion 
```
